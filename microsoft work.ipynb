## Create a new column in the DataFrame using List Comprehension - Direction


```python
# price difference =  close price of tommorow - close price of today
ms['Price1'] = ms['Close'].shift(-1)
ms.head()
```

```python
ms['Pricediff'] = ms['Price1']-ms['Close']
ms.head()
```

```python
# daily return = price difference / close price of today

ms['DailyrReturn']= ms['Pricediff']/ms['Close']
ms.head()
```

```python
ms['Direction'] = [1 if ms['Pricediff'].loc[ei]>0 else -1 for ei in ms.index ]
ms.head()
```

## Build a simple trading strategy

## Munging the stock data and add two columns - MA10 and MA50


```python
#import MS's stock data, add two columns - MA10 and MA50
#use dropna to remove any "NaN" data

ms['MA10'] = ms['Close'].rolling(10).mean()
ms['MA50'] = ms['Close'].rolling(50).mean()
ms = ms.dropna()
ms.head()
```

## Add "Shares" column to make decisions base on the strategy


```python
#Add a new column "Shares", if MA10>MA50, denote as 1 (long one share of stock), otherwise, denote as 0 (do nothing)

ms['Shares'] = [1 if ms.loc[ei, 'MA10']>ms.loc[ei, 'MA50'] else 0 for ei in ms.index]
ms.head()
```

## Add a new column "Profit" using List Comprehension

```python
# for any rows in ms, if Shares=1, the profit is calculated as the close price of 
#tomorrow - the close price of today. Otherwise the profit is 0.

#Plot a graph to show the Profit/Loss

ms['Close1'] = ms['Close'].shift(-1)
ms['Profit'] = [ms.loc[ei, 'Close1'] - ms.loc[ei, 'Close'] if ms.loc[ei, 'Shares']==1 else 0 for ei in ms.index]
ms['Profit'].plot()
plt.axhline(y=0, c='r');
```

## Use .cumsum() to display our model's performance if we follow the strategy


```python
#Use .cumsum() to calculate the accumulated wealth over the period

ms['wealth'] = ms['Profit'].cumsum()
ms.tail()
```

```python
#plot the wealth to show the growth of profit over the period

ms['wealth'].plot()
plt.title('Total money you win is {}'.format(ms.loc[ms.index[-2], 'wealth']));
```
